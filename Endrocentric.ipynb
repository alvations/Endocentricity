```python
import io, os
from collections import defaultdict
import operator
```

```python
domain2terms = defaultdict(list)

indir = 'terms_en/'
for filename in os.listdir('terms_en/'):
    domain = filename.rpartition('.')[0]
    with io.open(indir+filename, 'r', encoding='utf8') as fin:
        for line in fin:
            term = line.strip().split('\t')[1]
            domain2terms[domain].append(term)
```

```python
def s1(t1, terms):
    return set([t2 for t2 in terms if t1 != t2 and len(t1)>3 
                and t2 in t1]) 

def s2(t1, terms):
    return set([t2 for t2 in terms if t1 != t2 and len(t1)>3 and len(t2) > 2
                and (t1.startswith(t2 + ' ') if ' of ' in t1 else t1.endswith(' '+ t2))])
```

```python

hypohyper = {}
for domain in domain2terms:
    terms = domain2terms[domain]
    
    concato = {}
    for hypo in terms:
        concatenyms = s2(hypo, terms)
        if concatenyms:
            concato[hypo] = sorted(set(concatenyms), key=len)[-1]
            
    hypohyper[domain] = concato
```

```python
print hypohyper
```

```python

for domain in hypohyper:
    with io.open(domain+'.taxo', 'w') as fout:
        taxonomy = sorted(hypohyper[domain].items(), key=operator.itemgetter(1))
        for i, (hypo, hyper) in enumerate(taxonomy):
            fout.write(str(i) + '\t' + hypo + '\t' + hyper + '\n')

```
